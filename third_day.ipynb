```python
from string import ascii_lowercase
```

```python
lower = [s for s in ascii_lowercase]
upper = [s.upper() for s in ascii_lowercase]
```

```python
def _halfCut(incoming: str):
    half = int(len(incoming) /2)
    return incoming[:half], incoming[half:]

_halfCut('ABBA')
```

```python
from itertools import product

def iterate_over_both_rucksacks(line):
    for ll,rl in product(*_halfCut(line)):
        if ll == rl:
             return ll
```

```python
def score_lower(letter):
    return lower.index(letter) +1

def score_higher(letter):
    return upper.index(letter) + 27
```

```python
all_double_letters = [iterate_over_both_rucksacks(line) for line in _input]
```

```python
scores = [score_higher(letter) if letter.isupper() else score_lower(letter) for letter in all_double_letters]
```

```python
def batch(iterable, n=1):
    l = len(iterable)
    for ndx in range(0, l, n):
        yield iterable[ndx:min(ndx + n, l)]
```

```python
def iterate_over_three_rucksacks(r1, r2, r3):
        for i1, i2,i3 in product(r1,r2,r3):
            if i1==i2==i3:
                 return i1
```

```python
badges = [iterate_over_three_rucksacks(*triple) for triple in batch(_input,3)]
```

```python
scores_badges = [score_higher(letter) if letter.isupper() else score_lower(letter) for letter in badges]
```
