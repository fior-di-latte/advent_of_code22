```python
_stack = _input[:8]
```

```python
def _formatStack(stack: list)->list:
    _separated  =  [[line[i:i+3] for i in list(range(len(line)))[::4]]  for line in stack]
    _formated = [[None if l == '   ' else l for l in line] for line in _separated]
    return _formated
```

```python
_formatedStack = _formatStack(_stack)
```

```python
import re

txt = "move 15 from 8 to 9"
x = re.findall("\d{1,2}", txt)
```

```python
_instructions = [re.findall("\d{1,2}", line) for line in _input[10:]]
```

```python
_instructionsInts = [[int(i) for i in l] for l in _instructions]
```

```python
_stackCols = [[_formatedStack[line][stacknumber] for line  in range(len(_formatedStack))] for  stacknumber in range(9) ]
```

```python
_topUp = [list(reversed(i)) for i in _stackCols]
```

```python
_noNones = [[letter for letter in stack if letter ] for stack in _topUp]
_noNones
```

```python

def _move(n:int , fr:int, to: int, inputStacks=None):
    wipStacks = inputStacks.copy()
    _toBeAppended = list(reversed(wipStacks[fr-1][-n:]))
    wipStacks[fr-1] =     wipStacks[fr-1][:-n]
    wipStacks[to-1] =     wipStacks[to-1]+ _toBeAppended
    return wipStacks
    
```

```python
from functools import reduce
_res = reduce(lambda res, numbers: _move(*numbers, inputStacks=res), _instructionsInts, _noNones)
```

```python
''.join([stack[-1][1] for stack in _res])
```

```python

def _movep2(n:int , fr:int, to: int, inputStacks=None):
    wipStacks = inputStacks.copy()
    _toBeAppended = list(wipStacks[fr-1][-n:])
    wipStacks[fr-1] =     wipStacks[fr-1][:-n]
    wipStacks[to-1] =     wipStacks[to-1]+ _toBeAppended
    return wipStacks


_resp2 = reduce(lambda res, numbers: _movep2(*numbers, inputStacks=res), _instructionsInts, _noNones)
```
